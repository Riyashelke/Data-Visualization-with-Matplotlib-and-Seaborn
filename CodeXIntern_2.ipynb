```python
import pandas as pd

# Load the dataset
file_path = 'survey lung cancer.csv'
data = pd.read_csv(file_path)

# Display the first few rows and basic info about the dataset
data.head(), data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set up the visual style
sns.set(style="whitegrid")

# Plot 1: Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['AGE'], kde=True, color="skyblue")
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Gender distribution among lung cancer diagnoses
plt.figure(figsize=(8, 6))
sns.countplot(x='GENDER', hue='LUNG_CANCER', data=data, palette="Set2")
plt.title('Gender Distribution by Lung Cancer Diagnosis')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Lung Cancer')
plt.show()
```

```python
# Plot 3: Heatmap of correlations between numerical factors
plt.figure(figsize=(12, 10))
correlation_matrix = data.select_dtypes(include=['int64']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix of Factors')
plt.show()
```

```python
# Plot 4: Smoking vs. Lung Cancer
plt.figure(figsize=(8, 6))
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=data, palette="viridis")
plt.title('Smoking Levels by Lung Cancer Diagnosis')
plt.xlabel('Smoking Level')
plt.ylabel('Count')
plt.legend(title='Lung Cancer')
plt.show()
```

```python
# Plot 5: Pie chart for lung cancer diagnosis distribution
plt.figure(figsize=(8, 8))
lung_cancer_counts = data['LUNG_CANCER'].value_counts()
plt.pie(lung_cancer_counts, labels=lung_cancer_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightgreen'])
plt.title('Lung Cancer Diagnosis Distribution')
plt.show()
```

```python
# Plot 6: Line graph of age vs smoking levels
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="AGE", y="SMOKING", marker="o", color="blue")
plt.title('Smoking Levels Across Age')
plt.xlabel('Age')
plt.ylabel('Smoking Level')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Clean up column names by removing leading and trailing spaces
data.columns = data.columns.str.strip()

# Define the numeric features for box plots
numeric_features = ['AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 
                    'CHRONIC DISEASE', 'FATIGUE', 'WHEEZING', 'ALCOHOL CONSUMING', 
                    'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']

# Set up the figure
plt.figure(figsize=(15, 8))
sns.boxplot(data=data[numeric_features], orient='h', palette="Set3")
plt.title('Outlier Detection in Numerical Features')
plt.xlabel('Values')
plt.show()
```
